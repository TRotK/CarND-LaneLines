## Helper Functions

Below are some helper functions to help get you started. They should look familiar from the lesson!

````python
import math

def grayscale(img): 
    """Applies the Grayscale transform
    This will return an image with only one color channel
    but NOTE: to see the returned image as grayscale
    (assuming your grayscaled image is called 'gray')
    you should call plt.imshow(gray, cmap='gray')"""
    return cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    # Or use BGR2GRAY if you read an image with cv2.imread()
    # return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
def canny(img, low_threshold, high_threshold):
    """Applies the Canny transform"""
    return cv2.Canny(img, low_threshold, high_threshold)

def gaussian_blur(img, kernel_size):
    """Applies a Gaussian Noise kernel"""
    return cv2.GaussianBlur(img, (kernel_size, kernel_size), 0)

def region_of_interest(img, vertices):
    """
    Applies an image mask.
    
    Only keeps the region of the image defined by the polygon
    formed from `vertices`. The rest of the image is set to black.
    """
    #defining a blank mask to start with
    mask = np.zeros_like(img)   
    
    #defining a 3 channel or 1 channel color to fill the mask with depending on the input image
    if len(img.shape) > 2:
        channel_count = img.shape[2]  # i.e. 3 or 1 depending on your image
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255
        
    #filling pixels inside the polygon defined by "vertices" with the fill color    
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    
    #returning the image only where mask pixels are nonzero
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image


def draw_lines(img, lines, color=[255, 0, 0], thickness=2):
    """
    NOTE: this is the function you might want to use as a starting point once you want to 
    average/extrapolate the line segments you detect to map out the full
    extent of the lane (going from the result shown in raw-lines-example.mp4
    to that shown in P1_example.mp4).  
    
    Think about things like separating line segments by their 
    slope ((y2-y1)/(x2-x1)) to decide which segments are part of the left
    line vs. the right line.  Then, you can average the position of each of 
    the lines and extrapolate to the top and bottom of the lane.
    
    This function draws `lines` with `color` and `thickness`.    
    Lines are drawn on the image inplace (mutates the image).
    If you want to make the lines semi-transparent, think about combining
    this function with the weighted_img() function below
    """
    for line in lines:
        for x1,y1,x2,y2 in line:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

def hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with hough lines drawn.
    """
    lines = cv2.HoughLinesP(img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    line_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    draw_lines(line_img, lines)
    return line_img

# Python 3 has support for cool math symbols.

def weighted_img(img, initial_img, =0.8, =1., =0.):
    """
    `img` is the output of the hough_lines(), An image with lines drawn on it.
    Should be a blank image (all black) with lines drawn on it.
    
    `initial_img` should be the image before any processing.
    
    The result image is computed as follows:
    
    initial_img *  + img *  + 
    NOTE: initial_img and img must be the same shape!
    """
    return cv2.addWeighted(initial_img, , img, , )
````

## Build a Lane Finding Pipeline



Build the pipeline and run your solution on all test_images. Make copies into the `test_images_output` directory, and you can use the images in your writeup report.

Try tuning the various parameters, especially the low and high Canny thresholds as well as the Hough lines parameters.

````python
# TODO: Build your pipeline that will draw lane lines on the test_images
# then save them to the test_images_output directory.

# Compute line segments with the helper functions
input_dir = "test_images/"
for i in range(len(os.listdir(input_dir))):
    
    #reading in an image
    image = mpimg.imread(input_dir + os.listdir(input_dir)[i])

    # Grayscale the image
    gray = grayscale(image)

    # Define a kernel size and apply Gaussian smoothing
    kernel_size = 5
    blur_gray = gaussian_blur(gray, kernel_size)

    # Define parameters for Canny and apply
    low_threshold = 80
    high_threshold = 180
    edges = canny(blur_gray, low_threshold, high_threshold)

    # Define a four sided polygon to mask
    imshape = image.shape
    vertices = np.array([[(.15*imshape[1], imshape[0]), (.46*imshape[1], .6*imshape[0])
                          , (.54*imshape[1], .6*imshape[0]), (.95*imshape[1], imshape[0])]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)

    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 20     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 10 # minimum number of pixels making up a line
    max_line_gap = 10    # maximum gap in pixels between connectable line segments
    #line_image = np.copy(image)*0 # creating a blank to draw lines on

    # Run Hough on edge detected image
    # Output "line_image" is an image with hough lines drawn
    line_image = hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)

    # Output image with marked lanes
    output_image = weighted_img(line_image, image, =0.8, =1., =0.)

    # Display the output image
    plt.imshow(output_image)
    plt.show()

    # Write the output image into disk 
    output_dir = "test_images_output/line_segments/"
    os.makedirs(output_dir, exist_ok=True) # create a folder if it does not exist
    mpimg.imsave(output_dir + os.listdir(input_dir)[i], output_image)
````

````python
# Average/extrapolate the line segments

# Define linear regression function that models the left/right lane
def linear_regression(lines):
    """
    y = b0 + b1 * x.
    
    Returns parameters b0 and b1.
    """
    X, Y = [], []
    for line in lines:
        for x1,y1,x2,y2 in line:
            X.append(x1)
            Y.append(y1)
            X.append(x2)
            Y.append(y2)
            
    N = len(X)
    x_bar = sum(X) / N
    y_bar = sum(Y) / N
    
    numer = sum([xi * yi for xi, yi in zip(X, Y)]) - N * x_bar * y_bar
    denum = sum([xi**2 for xi in X]) - N * x_bar**2
     
    b1 = numer / denum
    b0 = y_bar - b1 * x_bar 
    
    return b0, b1


def my_draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """ 
    Separate line segments by their slope ((y2-y1)/(x2-x1)) and position
    to decide which segments are part of the left line vs. the right line.  
    Then, use linear regression to average the position of each of lines
    and extrapolate to the top and bottom of the lane.
    """
    # Classify two lanes according their slope and position
    lines_left, lines_right = np.empty(shape=(0, 4), dtype=np.int32), np.empty(shape=(0, 4), dtype=np.int32)
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2 - y1) / (x2 - x1)
            if slope < 0 and x1 <= img.shape[1] / 2 and x2 <= img.shape[1] / 2:
                lines_left = np.vstack((lines_left, line))
            elif slope > 0 and x1 > img.shape[1] / 2 and x2 > img.shape[1] / 2:
                lines_right = np.vstack((lines_right, line))
    
    # Reshape to match with "lines"
    lines_left = lines_left.reshape(len(lines_left), 1, 4)
    lines_right = lines_right.reshape(len(lines_right), 1, 4)
    
    # Use linear regression to fit lines on detected line segments
    b0_left, b1_left = linear_regression(lines_left)
    y1_hat = img.shape[0]
    x1_hat = (y1_hat - b0_left) / b1_left
    y2_hat = .6 * img.shape[0]
    x2_hat = (y2_hat - b0_left) / b1_left
    lanes = np.array([[x1_hat, y1_hat, x2_hat, y2_hat]], dtype=np.int32) 

    b0_right, b1_right = linear_regression(lines_right)
    y1_hat = img.shape[0]
    x1_hat = (y1_hat - b0_right) / b1_right
    y2_hat = .6 * img.shape[0]
    x2_hat = (y2_hat - b0_right) / b1_right
    lanes = np.vstack((lanes, np.array([[x1_hat, y1_hat, x2_hat, y2_hat]], dtype=np.int32)))
    
    lanes = lanes.reshape(len(lanes), 1, 4) # lanes.shape = (2, 1, 4)
    
    for lane in lanes:
        for x1,y1,x2,y2 in lane:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

            
# Define my own hough lines function that outputs averaged/extrapolated lane marks
def my_hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with averaged/extrapolated hough lines drawn.
    """
    lines = cv2.HoughLinesP(
        img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    
    # Draw lanes on image
    lane_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    my_draw_lines(lane_img, lines, thickness=10) # calls my own draw_lines function to draw lanes
    return lane_img


input_dir = "test_images/"
for i in range(len(os.listdir(input_dir))):
    
    #reading in an image
    image = mpimg.imread(input_dir + os.listdir(input_dir)[i])

    # Grayscale the image
    gray = grayscale(image)

    # Define a kernel size and apply Gaussian smoothing
    kernel_size = 5
    blur_gray = gaussian_blur(gray, kernel_size)

    # Define parameters for Canny and apply
    low_threshold = 80
    high_threshold = 180
    edges = canny(blur_gray, low_threshold, high_threshold)

    # Define a four sided polygon to mask
    imshape = image.shape
    vertices = np.array([[(.15*imshape[1], imshape[0]), (.46*imshape[1], .6*imshape[0])
                          , (.54*imshape[1], .6*imshape[0]), (.95*imshape[1], imshape[0])]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)

    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 2 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 20     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 10 # minimum number of pixels making up a line
    max_line_gap = 10    # maximum gap in pixels between connectable line segments
    #line_image = np.copy(image)*0 # creating a blank to draw lines on

    # Run my own Hough on edge detected image
    # Output "lane_image" is an image with left and right lane drawn
    lane_image = my_hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)

    # Output image with marked lanes
    output_image = weighted_img(lane_image, image, =0.8, =1., =0.)

    # Display the output image
    plt.imshow(output_image)
    plt.show()

    # Write the output image into disk
    output_dir = "test_images_output/lane_marks/"
    os.makedirs(output_dir, exist_ok=True) # create a folder if it does not exist
    mpimg.imsave(output_dir + os.listdir(input_dir)[i], output_image)
````

````python
# Pipeline for challenge.mp4
from sklearn import linear_model

# Define linear regression function with RANSAC that rejects outliers and models the left/right lane
def ransac_regression(lines):
    """
    y = b0 + b1 * x.
    
    Returns parameters b0 and b1.
    """
    X, Y = [], []
    for line in lines:
        for x1,y1,x2,y2 in line:
            X.append(x1)
            Y.append(y1)
            X.append(x2)
            Y.append(y2)
            
    X = np.array([X]).T # shape [n_samples, n_features]
    Y = np.array(Y)

    # Robustly fit linear model with RANSAC algorithm
    model_ransac = linear_model.RANSACRegressor(linear_model.LinearRegression())
    model_ransac.fit(X, Y)
    
    b0 = model_ransac.estimator_.intercept_
    b1 = model_ransac.estimator_.coef_[0]
    inlier_mask = model_ransac.inlier_mask_
    y_min = Y[inlier_mask].min(axis=0)

    return b0, b1, y_min


# Re-define my own draw_lines function using RANSAC algorithm
def my_draw_lines(img, lines, color=[255, 0, 0], thickness=10):
    """ 
    Separate line segments by their slope ((y2-y1)/(x2-x1)) and position
    to decide which segments are part of the left line vs. the right line.  
    Then, use linear regression to average the position of each of lines
    and extrapolate to the top and bottom of the lane.
    """
    # Classify two lanes according their slope and position
    lines_left, lines_right = np.empty(shape=(0, 4), dtype=np.int32), np.empty(shape=(0, 4), dtype=np.int32)
    for line in lines:
        for x1,y1,x2,y2 in line:
            slope = (y2 - y1) / (x2 - x1)
            if slope < -.5 and x1 <= .55*img.shape[1] and x2 <= .55*img.shape[1]:
                lines_left = np.vstack((lines_left, line))
            elif slope > .5 and x1 > .45*img.shape[1] and x2 > .45*img.shape[1]:
                lines_right = np.vstack((lines_right, line))
    
    # Reshape to match with "lines"
    lines_left = lines_left.reshape(len(lines_left), 1, 4)
    lines_right = lines_right.reshape(len(lines_right), 1, 4)
    
    # Use linear regression to fit lines on detected line segments
    lanes = np.empty(shape=(0, 4), dtype=np.int32)
    if lines_left.size: # skip if no line detected
        b0_left, b1_left, y2_hat = ransac_regression(lines_left)
        y1_hat = img.shape[0]
        x1_hat = (y1_hat - b0_left) / b1_left
#         y2_hat = lines_left.min(axis=0)[:, [1, 3]].min(axis=1)[0]
        y2_hat = .62 * img.shape[0] # comment this for adaptive extrapolation
        x2_hat = (y2_hat - b0_left) / b1_left
        lanes = np.array([[x1_hat, y1_hat, x2_hat, y2_hat]], dtype=np.int32) 

    if lines_right.size: # skip if no line detected
        b0_right, b1_right, y2_hat = ransac_regression(lines_right)
        y1_hat = img.shape[0]
        x1_hat = (y1_hat - b0_right) / b1_right
#         y2_hat = lines_right.min(axis=0)[:, [1, 3]].min(axis=1)[0]
        y2_hat = .62 * img.shape[0] # comment this for adaptive extrapolation
        x2_hat = (y2_hat - b0_right) / b1_right
        lanes = np.vstack((lanes, np.array([[x1_hat, y1_hat, x2_hat, y2_hat]], dtype=np.int32)))
    
    lanes = lanes.reshape(len(lanes), 1, 4) # lanes.shape = (2, 1, 4)
    
    for lane in lanes:
        for x1,y1,x2,y2 in lane:
            cv2.line(img, (x1, y1), (x2, y2), color, thickness)

            
# Define my own hough lines function that outputs averaged/extrapolated lane marks
def my_hough_lines(img, rho, theta, threshold, min_line_len, max_line_gap):
    """
    `img` should be the output of a Canny transform.
        
    Returns an image with averaged/extrapolated hough lines drawn.
    """
    lines = cv2.HoughLinesP(
        img, rho, theta, threshold, np.array([]), minLineLength=min_line_len, maxLineGap=max_line_gap)
    
    # Draw lanes on image
    lane_img = np.zeros((img.shape[0], img.shape[1], 3), dtype=np.uint8)
    if lanes.size: # skip if no lanes detected
        draw_lines(lane_img, lanes, thickness=10) # draw lanes
    
    return lane_img


def process_image(image):
    # NOTE: The output you return should be a color image (3 channel) for processing video below
    # TODO: put your pipeline here,
    # you should return the final output (image where lines are drawn on lanes)

    # Grayscale the image
#     gray = grayscale(image) # skip grayscale for better contrast

    # Define a kernel size and apply Gaussian smoothing
    kernel_size = 5
    blur_gray = gaussian_blur(image, kernel_size)

    # Define parameters for Canny and apply
    low_threshold = 100
    high_threshold = 200
    edges = canny(blur_gray, low_threshold, high_threshold)

    # Define a four sided polygon to mask
    imshape = image.shape
    vertices = np.array([[(.15*imshape[1], imshape[0]), (.45*imshape[1], .6*imshape[0])
                          , (.55*imshape[1], .6*imshape[0]), (.95*imshape[1], imshape[0])]], dtype=np.int32)
    masked_edges = region_of_interest(edges, vertices)
    
    # Define the Hough transform parameters
    # Make a blank the same size as our image to draw on
    rho = 1 # distance resolution in pixels of the Hough grid
    theta = np.pi/180 # angular resolution in radians of the Hough grid
    threshold = 20     # minimum number of votes (intersections in Hough grid cell)
    min_line_length = 10 # minimum number of pixels making up a line
    max_line_gap = 15    # maximum gap in pixels between connectable line segments
    #line_image = np.copy(image)*0 # creating a blank to draw lines on

    # Run my own Hough on edge detected image
    # Output "lane_image" is an image with left and right lane drawn
#     lane_image = hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)
    lane_image = my_hough_lines(masked_edges, rho, theta, threshold, min_line_length, max_line_gap)
    
    # Output image with marked lanes
    result = weighted_img(lane_image, image, =0.8, =1., =0.)
    
    return result
````
